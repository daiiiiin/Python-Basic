# Step 10. states dataframe에는 각 주의 Code, abbreviation, fullname이 포함되어 있습니다. states와 baby_names dataframe을 적절히 병합하여 baby_names 내 각 주의 코드명을 fullname으로 변경하세요.

```python
baby_names = pd.read_csv('US_Baby.csv')
```

```python
baby = baby_names.drop(labels = ['Unnamed: 0', 'Id'], axis=1)
```

```python
baby_names = pd.merge(baby, states, left_on = 'State', right_on='Code').drop(['Code', 'Abbrev'], axis=1)
```

```python
baby_names = baby_names.drop(labels = 'State_x', axis=1)
baby_names.columns = ["Name", "Year", "Gender", "Count", "State"]
```

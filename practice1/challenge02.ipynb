## 0. 클래스 생성 
- 빈 사각형, 꽉 찬 사각형 출력하는 클래스(Box) 구현
- 생성자 : 사각형 가로 길이(width), 사각형 세로 길이 (length)
- 메소드 : 꽉 찬 사각형(black_box), 빈 사각형(white_box)
- black_box, white_box의 출력 예시는 아래와 같음
- 세로 길이 3, 가로 길이 7 일 때 예시임

```python
class make_box(object):
    def __init__(self, width, length):
        self.width = width
        self.length = length
        
    def black_box(self, width, length):
        for i in range(length):
            print('*'*width)
    
    def white_box(self, width, length):
        for i in range(0, length):
            if i == 0 or i == (length-1):
                print('*'*width)
            else:
                print('*' + ' '*(width-2) + '*')
```

```python
box = make_box(7, 3)
box.black_box(7, 3)
```

```python
box.white_box(7, 3)
```

```python
# 예시
box.black_box()
```

```python
# 예시
box.white_box()
```

## 8.  새로운 컬럼 추가 1
- 미세먼지 농도별 예보 등급을 기준으로, 
- pm10 농도 '좋음'에 해당하는 값들은 'good','보통'의 경우 'soso', '나쁨'의 경우 'bad', '매우 나쁨'의 경우 
'very bad'의 값을 가지는 컬럼(컬럼명 : 'new_pm10')을 새로 추가
- pm2.5 농도 '좋음'에 해당하는 값들은 'good','보통'의 경우 'soso', '나쁨'의 경우 'bad', '매우 나쁨'의 경우 
'very bad'의 값을 가지는 컬럼(컬럼명 : 'new_pm2.5')을 새로 추가

```python
new_pm10 = []
for i in df['hour_bef_pm10']:
    if (i>=0) & (i<=30):
        new_pm10.append('good')
    elif (i>=31) & (i<=80):
        new_pm10.append('soso')
    elif (i>=81) & (i<=150):
        new_pm10.append('bad')
    else:
        new_pm10.append('very bad')
```

```python
new_pm25 = []
for i in df['hour_bef_pm2.5']:
    if (i>=0) & (i<=15):
        new_pm25.append('good')
    elif (i>=16) & (i<=35):
        new_pm25.append('soso')
    elif (i>=36) & (i<=75):
        new_pm25.append('bad')
    else:
        new_pm25.append('very bad')
```

```python
df['new_pm10'] = new_pm10
df['new_pm2.5'] = new_pm25
```
